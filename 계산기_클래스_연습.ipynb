```python
# 1. 숫자를 리스트로 묶기 / 차례대로 꺼내야 하기 때문에

nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')

for i in nums1:
    if isinstance(i,int):                # 숫자인지 확인
        print(i)
    elif isinstance(i,(list,tuple)):     # 타입을 여러개 확일 할 경우
        for j in i:
            if isinstance(j,int):
                print(j)
```

```python
# 2. 값 저장소 만들기

nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')
numbers = []                    # 숫자를 저장할 곳

for i in nums1:
    if isinstance(i,int):
        numbers.append(i)
    elif isinstance(i,(list,tuple)):
        for j in i:
            if isinstance(j,int):
                numbers.append(j)   # 값 저장해주기
print(numbers)
```

```python
# 3. 함수만들기 / 리턴값 주기

# nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')

def get_numbers(nums1):
    numbers = []

    for i in nums1:
        if isinstance(i,int):
            numbers.append(i)
        elif isinstance(i,(list,tuple)):
            for j in i:
                if isinstance(j,int):
                    numbers.append(j)

    return numbers
```
