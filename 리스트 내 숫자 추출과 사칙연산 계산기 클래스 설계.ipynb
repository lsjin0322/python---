# [문제]
- 리스트 안에 숫자와 문자가 섞여 있다
- 숫자와 문자를 구분해야한다
- 숫자만 추출해서 계산하는 클래스만들기

# [해결방식]
1. 리스트에서 숫자와 문자를 분리한다.
2. 숫자는 리스트로 묶기 / 차례대로 꺼내야 하기 때문에
3. 계산은 숫자 리스트를 기준으로 진행한다


```python
# 1. 숫자와 문자를 먼저 구분하기

nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')

for i in nums1:
    if isinstance(i,int):                # 숫자인지 확인
        print(i)
    elif isinstance(i,(list,tuple)):     # 타입이 여러개일때 사용하여 숫자만 출력하도록 이중for문 사용하기
        for j in i:
            if isinstance(j,int):
                print(j)

for i in nums1:
    if isinstance(i,str):                 # 문자인지 확인
        print(i)
    elif isinstance(i,(list,tuple)):
        for k in i:
            if isinstance(k,str):
                print(k)
```

```python
# 2. 값 저장소 만들기

nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')
numbers = []                    # 숫자를 저장할 곳
strings = []                    # 문자를 저장할 곳

for i in nums1:
    if isinstance(i,int):
        numbers.append(i)
    if isinstance(i,str):
        strings.append(i)

    elif isinstance(i,(list,tuple)):
        for j in i:
            if isinstance(j,int):   # 만약에 j가 숫자라면
                numbers.append(j)   # 숫자만 분리한 값을 저장하기

            elif isinstance(j,str):  # 위의 조건이 아니라 문자라면
                strings.append(j)    # 문자만 분리한 값을 저장하기

print(numbers)
print(strings)
```

```python
# 3. 함수만들기 / 리턴값 주기

# nums1 = (1,2,3,4,[1,2,3, 'a'],(1,2,3),'a','a')

def get_numbers(nums1):
    numbers = []
    strings = []

    for i in nums1:
        if isinstance(i,int):
            numbers.append(i)
        elif isinstance(i,(list,tuple)):
            for j in i:
                if isinstance(j,int):
                    numbers.append(j)

    for i in nums1:
        if isinstance(i,str):
            strings.append(i)
        elif isinstance(i,(list,tuple)):
            for k in i:
                if isinstance(k,str):
                    strings.append(k)



    return numbers, strings
```

```python
# 4. 클래스로 옮기기

# def get_numbers(nums1):
#     numbers = []



class NumberExtractor:
    def __init__(self):
        self.numbers = []                         # 숫자와 문자열을 저장할 리스트를 클래스 속성에 만들기
        self.strings = []

    def get_numbers(self,nums1):                   # 메소드
            self.numbers = []                      # 이전 데이터가 남지 않도록 초기화
            self.strings = []


            for i in nums1:
                if isinstance(i,int):              # 입력값 하나씩 확인
                    self.numbers.append(i)         # 숫자인 경우 numbers에 저장하기

                elif isinstance(i,(list,tuple)):   # 리스트나 튜플일 경우 안쪽 값도 확인하기
                    for j in i:
                        if isinstance(j,int):      # 내부요소가 숫자라면 저장하기
                            self.numbers.append(j)
                        elif isinstance(j,str):     # 내부요소가 문자열이라면 저장하기
                            self.strings.append(j)

                elif isinstance(i,str):             # 문자열인 경우 strings에 저장하기
                    self.strings.append(i)

            return self.numbers                     # 숫자 리스트로 반환
```

```python
# 5. 계산기능만 있는 클래스 만들기

class calculator:
    def __init__(self,nums1):
        self.numbers = nums1       # 모든 함수에서 사용가능

    def add (self):                # 더하기 시작은 0
        result = 0

        for n in self.numbers:     # 리스트 하나씩 꺼내서 처리
            result = result + n
        return result

    def sub(self):
        result = self.numbers[0]

        for n in self.numbers[1:]:
            result = result - n
        return result

    def mul(self):
        result = 1
        for n in self.numbers:
            result = result * n
        return result

    def div(self):
        result = self.numbers[0]
        for n in self.numbers[1:]:
            result /= n
        return result

    def avg(self):
        return self.add() / len(self.numbers)     # 만들어놓았던 add 기능 다시 쓰기

    def get_max(self):
        result = self.numbers[0]
        for n in self.numbers:
            if n > result:
                result = n
        return result

    def get_min(self):
        result = self.numbers[0]
        for n in self.numbers:
            if n < result:
                result = n
        return result
```
