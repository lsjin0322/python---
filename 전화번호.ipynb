```python
def filter_numbers(data):
    result = []
    for x in data:
        try:
            result.append(float(x))
        except:
            continue
    return result


def total(data):
    nums = filter_numbers(data)
    return sum(nums)


def average(data):
    nums = filter_numbers(data)
    if len(nums) == 0:
        return 0
    return sum(nums) / len(nums)


def maximum(data):
    nums = filter_numbers(data)
    if len(nums) == 0:
        return None
    return max(nums)


def minimum(data):
    nums = filter_numbers(data)
    if len(nums) == 0:
        return None
    return min(nums)


# 사용 예시
data = [1, 2, "a", 5, "10"]

print(total(data))
print(average(data))
print(maximum(data))
print(minimum(data))
```

```python
# 전화번호 프로그램
# 조건) 클래스 상속 사용하기
# 사용자 정보 저장 및 검색 기능 만들기


# 부모클래스
class Contact:
    def __init__(self, name,phone):
        self.name =name
        self.phone = phone

# 자식 클래스
class Person(Contact):
    def __init__(self,name,phone,email,company):
        self().__init__(name,phone)
        self.email= email
        self.company = company

# 리스트 저장공간
people = []

# 저장
p1 = Person("철수", "010-1234-5678","aaa@naver.com", "카카오")
people.append(p1)

# 검색
search_name = "철수"

for p in people:
    if p.name == search_name:
        print("이름:" , p.name)
        print("전화번호:", p.phone)
        print("이메일:", p.email)
        print("회사:", p.company)
```
